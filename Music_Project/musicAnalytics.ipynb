```python
# Fetch a user's top 20 artists (overall)
user = network.get_user(USERNAME)
top_artists_user_overall = user.get_top_artists(limit=20, period="overall")

# Create a list of dictionaries for the DataFrame
artist_data = [
    {
        'Rank': i + 1,
        'Artist': artist.item.name,
        'Playcount': artist.weight,
        'Last.fm URL': artist.item.get_url()
    }
    for i, artist in enumerate(top_artists_user_overall)
]

# Create the DataFrame
df = pd.DataFrame(artist_data)

# Display the DataFrame
df
```
